```python

import talib
from talib import EMA, DEMA

from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA, GOOG

class EmaCross(Strategy):
    
    def init(self):
        
        Close = self.data.Close
        
        self.ema20 = self.I(EMA, Close, timeperiod=20)
        
        self.ema200 = self.I(EMA, Close, timeperiod=200)
        
        #self.crossover = crossover(self.ema20, self.ema200)
        
        #self.signal_add(crossover)
        
    def next(self):
        
        if (self.data.Close > self.ema20) and (self.ema20 > self.ema200) and ((abs(self.ema20 - self.ema200)/self.ema20)*100 > 0.5):
            self.buy()
            
        elif (self.data.Close < self.ema20) and (self.ema20 < self.ema200) and ((abs(self.ema20 - self.ema200)/self.ema20)*100 > 0.5):
            self.sell()
        
        

bt = Backtest(GOOG, EmaCross, cash=10000, commission=.002)
bt.run()
bt.plot()
```
